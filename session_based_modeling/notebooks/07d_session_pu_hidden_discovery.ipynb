# 07d. PU-Learning Hidden Candidate Discovery

KMeans cluster 1을 확정 정답이 아니라 `known positive seed`로만 사용하고, KMeans cluster 0은 negative가 아니라 `unlabeled`로 취급하는 PU-learning 관점의 hidden candidate discovery 노트북이다.

## 하는 일

- `cluster_number == 1`을 known positive로 정의
- `cluster_number == 0`을 normal이 아니라 unlabeled로 정의
- conservative normal seed와 reliable negative seed 생성
- reliable negative를 반복 샘플링하는 bagging PU 학습
- Logistic / Random Forest / Gradient Boosting PU classifier 반복 학습
- 각 session이 known positive와 얼마나 닮았는지 `pu_positive_score` 계산
- KMeans cluster 0 안에서 `pu_hidden_candidate`와 `pu_watchlist` 분리
