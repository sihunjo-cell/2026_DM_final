# 07c. KMeans-Independent Hidden Candidate Discovery

KMeans 결과를 정답 라벨로 쓰는 한계를 줄이기 위한 session-level hidden candidate discovery 노트북이다.

이 파일은 `cluster_number == 1`을 학습 목표로 사용하지 않는다. KMeans cluster는 최종 후보를 해석할 때만 사용한다.

## 하는 일

- 확실히 정상으로 보이는 normal seed session 정의
- normal-only Isolation Forest / LOF / One-Class SVM / PCA reconstruction anomaly score 계산
- GMM low-density 및 cluster ambiguity score 계산
- HDBSCAN noise flag 계산
- rule mismatch score와 독립 anomaly evidence를 합쳐 `unsupervised_hidden_score` 생성
- KMeans cluster와 독립 점수의 관계로 후보를 네 그룹으로 분리

후보 그룹:

- `confirmed_suspicious`: KMeans cluster 1이고 독립 점수도 높음
- `hidden_candidate`: KMeans cluster 0이지만 독립 점수가 높음
- `kmeans_only_candidate`: KMeans cluster 1이지만 독립 점수는 낮음
- `normal_low_priority`: KMeans cluster 0이고 독립 점수도 낮음
